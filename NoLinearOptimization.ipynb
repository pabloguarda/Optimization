# Optimization in problems with a single decision variable


## Define function


```r
objfunc = function(x){2*x^2+1000*4/x}
```

## Plot function

```r
options(jupyter.plot_mimetypes = 'image/png')

plot(x=-100:100, y= objfunc(-100:100), type='l')
```

## Newton-Rhapson

### First and second derivative of the function

```r
diff1 <- function(x){4*x-4000/x^2}
diff2 <- function(x){4+8000/x^3}
```

### Next point in the iteration

```r
nextXNewton <- function(x){x-x*(x^3-1000)/(x^3+2000)}
```

### Convergency criterion

```r
convergency <- function(i1,i2){abs((objfunc(i2)-objfunc(i1)))}
```

### Iterations

```r
#Initial point for the iterations
Newton0 <- -10
```

```r
#First iteration
Newton1 <- nextXNewton(Newton0)
print(paste("First iteration: ",Newton1))
print(paste("Convergency: ",convergency(Newton0,Newton1)))
```

```r
#Second iteration
Newton2 <- nextXNewton(Newton1)
print(paste("Second iteration: ",Newton2))
print(paste("Convergency: ",convergency(Newton1,Newton2)))
```

```r
#Third iteration
Newton3 <- nextXNewton(Newton2)
print(paste("Third iteration: ",Newton3))
print(paste("Convergency: ",convergency(Newton2,Newton3)))
```

```r
#Fourth iteration
Newton4 <- nextXNewton(Newton3)
print(paste("Fourth iteration: ",Newton4))
print(paste("Convergency: ",convergency(Newton3,Newton4)))
```

```r
#Fifth iteration
Newton5 <- nextXNewton(Newton4)
print(paste("Fifth iteration: ",Newton5))
print(paste("Convergency: ",convergency(Newton4,Newton5)))
```
